## 3 · The AI Load Question

August 2025 average dispatch by hour of day (24 stacked-area columns), with a flat blue slab grown on top representing **always-on AI data center load**. The slab grows from 0 → 25 GW.

### Why this framing

AI training and inference workloads are **flat 24/7** — unlike crypto mining, they don't curtail when prices spike (latency requirements, depreciation pressure on $50M+ accelerator clusters). They sit on top of the existing demand curve as a constant block. That single property reshapes the grid in a way every public figure on the topic has been understating.

### The early trends in numbers

- **ERCOT's 2024 record peak demand: ~85.5 GW** (Aug 20, 2023 had been 85.4; 2024 broke it again).
- **Large-load interconnection queue: tens of GW from data centers**, the bulk filed in 2023–2024. Crypto mining was the precursor (~2 GW); AI is an order of magnitude bigger.
- **Texas SB 6 (2025)** gave ERCOT new authority over interconnecting large loads (>75 MW) precisely because of this surge.
- **Stargate (Abilene)** — OpenAI/Oracle campus — alone is ~1.2 GW announced. Meta, Google, Microsoft, xAI all building in TX.
- ERCOT's high-scenario 2030 forecast pushes peak demand toward **150 GW**, nearly doubling 2024.

### What the slab does to the duck

Adding 25 GW flat on top of an average August day (~57 GW base) is a ~44% load increase. Because the slab is constant:

1. **Daytime** the new load is partially absorbed by the solar peak — solar's value goes *up*, not down.
2. **Nighttime** the new load can only be served by gas, nuclear, coal, or batteries discharging energy that was charged during the day. Wind helps when it's blowing; on a calm August night, it doesn't.
3. **The duck stops deepening.** AI load fills in the belly. The grid's flexibility problem partially *solves itself* — but only if you keep building dispatchable capacity. Otherwise the belly turns into a cliff.

## Takeaways for the poster / essay

**Three story arcs the data already supports:**

1. **Solar is the year's biggest mover.** +40% YoY, deepens the duck month by month, becomes the majority of midday generation in the summer panels.
2. **Post-Uri Texas held.** The peak week was severe but the grid never broke — gas + coal carried it. This is a *resilience* story, not a *failure* story, and it's underreported.
